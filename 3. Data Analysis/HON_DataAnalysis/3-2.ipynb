# 누락된 값 처리하기

### np.nan: 누락된 값으로 표시하기

```python
# 도서권수 열의 첫 행 값을 None으로 바꾼후 누락된 값 세기
ns_book4.loc[0, '도서권수'] = None
ns_book4['도서권수'].isna().sum()
```

### fillna(): 누락된 값 바꾸기(1)

```python
# 누락된 값을 찾아 불리언으로 반환
set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
# 누락된 값을 빈 문자열로 바꾼다.
ns_book4. loc[set_isbn_na_rows, '세트 ISBN'] = ''
# 누락된 값이 몇 개인지 센다.
ns_book4['세트 ISBN'].isna().sum()
```

```python
# 더 편리한 방법
# fillna()에 원하는 값을 전달하면 NaN을 대체하고 새로운 데이터프레임을 반환함.
# isna()가 NaN 개수를 센다.
ns_book4.fillna('없음').isna().sum()
```

```python
# 특정 열만 선택해서 NaN 바꾸기
ns_book4['부가기호'].fillna('없음').isna().sum()
```

```python
# 특정 열의 NaN을 바꾸면서 전체 프레임 반환하려면 딕셔너리로 전달하기.
ns_book4.fillna({'부가기호':'없음'}).isna().sum()
```

### replace(): 누락된 값 바꾸기(2)

```python
# 1. 바꾸려는 값이 하나일 때
# replace(원래값, 새로운값)
ns_book4.replace(np.nan, '없음').isna().sum()
```

```python
# 2. 바꾸려는 값이 여러 개일 때
# replace([원래값1, 원래값2],[새로운값1, 새로운값2])
# NaN을 없음으로, 2021을 21로
ns_book4.replace([np.nan,'2021'], ['없음','21']).head(2)
# replace({원래값1: 새로운값1, 원래값2:새로운값2})
# ns_book4.replace({np.nan: '없음', '2021':'21'}).head(2) 딕셔너리 형식으로도 가능
```

```python
# 3. 열마다 다른 값으로 바꿀 때
# replace({열이름:원래값}, 새로운 값)
ns_book4.replace({'부가기호':np.nan},'없음').head(2)
```

```python
# replace({열이름:{원래값1:새로운 값1}}) 으로도 같은 결과.
ns_book4.replace({'부가기호':{np.nan:'없음'}, '발행년도':{'2021':'21'}}).head(2)
```

### 숫자 찾기:/d

```python
# 뒤에 두자리만 그룹으로 묶을 때는 \d\d(\d\d)
# 패턴 안에 있는 첫 번째 그룹 추출 \1
ns_book4.replace({'발행년도':{r'\d\d(\d\d)':r'\1'}}, regex=True)[100:102]
```

```python
# 정규표현식이 반복될 때는 중괄호 사용해서 개수를 지정할 수 있음.
ns_book4.replace({'발행년도':{r'\d{2}(\d{2})':r'\1'}}, regex=True)[102:104]
```

### 문자 찾기: 마침표(.)
- (지은이)라는 문자열 그대로를 찾아야 합니다. 즉, 괄호를 특수한 의미로 사용하는 것이 아니라 문자 그대로 괄호를 찾아야 하는 상황입니다. 이럴 때는 괄호 앞에 역슬래시(\)를 붙여서 괄호를 문자 그대로 인식하도록 합니다. 이를 '이스케이프(escape)'라고 합니다.
- r'\1\2'는 정규 표현식의 교체 부분입니다. 여기서 '\1'과 '\2'는 각각 첫 번째와 두 번째 괄호((.*))에 의해 찾아진 부분을 참조합니다.

```python
ns_book4.replace({'저자':{r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)':r'\1\2'},
                  '발행년도':{r'\d{2}(\d{2})':r'\1'}}, regex=True)[100:102]
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
    url = 'https://www.yes24.com/product/search?query={}'
    r = requests.get(url.format(isbn))
    soup = BeautifulSoup(r.text, 'html.parser')
    title = soup.find('a', attrs={'class':'gd_name'}).get_text()
    return title
```

```python
import re
```

re 모듈의 findall() 함수를 사용하면 원하는 정규식에 매칭되는 모든 문자열을 찾아 리스트로 반환해준다.
첫번째 매개변수는 원하는 정규식이고, 두 번째 매개변수는 검색 대상 문자열이다.

```python
def get_book_info(row):
    title = row['도서명']
    author = row['저자']
    pub = row['출판사']
    year = row['발행년도']

    url = 'https://www.yes24.com/product/search?query={}'
    r = requests.get(url.format(row['ISBN']))
    soup = BeautifulSoup(r.text, 'html.parser')

    try:
        if pd.isna(title):
            title = soup.find('a', attrs={'class':'gd_name'}).get_text()
    except AttributeError:
        pass

    try:
        if pd.isna(author):
            authors = soup.find('span', attrs={'class':'authPub info_auth'}).find_all('a')
            author_list = [auth.get_text() for auth in authors]
            author = ','.join(author_list)
    except AttributeError:
        pass

    try:
        if pd.isna(pub):
            pub = soup.find('span', attrs={'class':'authPub info_pub'}).find('a').get_text()
    except AttributeError:
        pass

    try:
        if year == -1:
            year_str = soup.find('span', attrs={'class':'authPub info_date'}).get_text()
            year = re.findall(r'\d{4}', year_str)[0]
    except AttributeError:
        pass

    return title, author, pub, year
```
