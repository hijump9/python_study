# 출판사별 발행 도서 개수 산점도 그리기

```python
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

### 고유한 출판사 목록 만들기

```python
top30_pubs = ns_book7['출판사'].value_counts()[:30]
top30_pubs
```

```python
# 상위 30개 출판사에 해당하는 행을 표시하는 불리언 인덱스 만들기
top30_pubs_idx = ns_book7['출판사'].isin(top30_pubs.index) # 상위 30개에 해당하는 출판사는 True, 아닌 출판사는 False 반환
top30_pubs_idx
```

```python
# 불리언 배열에서 True인 원소의 개수 세기
top30_pubs_idx.sum()
```

```python
# 1000개만 랜덤으로 뽑아보자
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
```

## 산점도 그리기

```python
from matplotlib.font_manager import findSystemFonts
findSystemFonts()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
# 값에 따라 마커 크기를 다르게 나타내기
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], s=ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
# 마커꾸미기
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], linewidth=0.5, edgecolor='k',
           alpha=0.3, s=ns_book8['대출건수']*2, c=ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
# 컬러맵
fig, ax = plt.subplots(figsize=(10, 8))
sc = ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], linewidth=0.5, edgecolor='k',
           alpha=0.3, s=ns_book8['대출건수']**1.3, c=ns_book8['대출건수'], cmap='jet')
ax.set_title('출판사별 발행 도서')
fig.colorbar(sc)
fig.show()
```
