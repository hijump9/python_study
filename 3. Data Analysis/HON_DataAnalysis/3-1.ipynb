# 중복된 행 찾기: duplicated()
- 중복된 행 중에서 처음 행을 제외한 나머지행은 True, 그 외의 중복되지 않은 나머지 모든 행은 False로 표시.
- 도서명이 중복되어서 기록되어서 한 책의 대출 건수가 분리되서 기록되는 경우가 있다.
- 한 책의 대출 건수를 모두 구하기 위해 중복된 행을 찾는다.

```python
sum(ns_book.duplicated()) # True를 1로 인식.
```

```python
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
```

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False) # 중복된 행을 모두 True로 표시한 불리언 배열 반환
ns_book3 = ns_book[dup_rows]
ns_book3.head(10)
```

파친코는 두 권짜리 도서라서 '권'열에 1과 2로 표시되어있고, 보건교사 안은영은 한 권짜리 도서인데, 인기가 많아 도서관에서 두 권을 구입한 것으로 추정.

# 그룹별로 모으기
- 파친코가 두 행이 있지만 하나의 파친코로 대출건수를 구하고 싶다.
- 따라서 그룹으로 묶을 기준 열 4개를 선택해서 같은 책의 대출건수를 하나로 합친다. (열 4개만 고른 이유는 출력 화면이 간단하게 하기위함)

```python
count_df = ns_book[['도서명','저자','ISBN', '권','대출건수']]
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False)
loan_count = group_df.sum()
```

```python
count_df.head()
```

```python
count_df = ns_book[['도서명','저자','ISBN', '권','대출건수']]
loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum() # NaN이 있는 행을 삭제하지 않음.
```

도서명, 저자, ISBN, 권 까지 인덱스이고, 각 책의 대출건수를 더한 결과가 대출건수 열에 저장된다.

# 원본데이터 업데이트하기
대출건수를 원본 데이터프레임에 업데이트해야하는데,
원본 데이터프레임에는 중복된 데이터가 있으므로 업데이트 하기 전에
1. duplicated() 메서드로 중복된 행을 True 로 표시한 불리언 배열 만들기
2. 1번에서 구한 불리언 배열을 반전시켜서 중복되지 않은 고유한 행을 True로 표시
3. 2번에서 구한 불리언 배열을 사용해 원본 배열에서 고유한 행만 선택

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권']) # 중복된 행을 True로 표시
unique_rows = ~dup_rows # 고유한 행을 True로 표시
ns_book3 = ns_book[unique_rows].copy() # 고유한 행만 선택
```

```python
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))
```

## 원본 데이터프레임 인덱스 설정하기

```python
# 지정한 열을 인덱스로 설정할 때 set_index 사용
ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
ns_book3.head()
```

## 업데이트하기: update() 메서드

```python
ns_book3.update(loan_count)
ns_book3.head()
```

```python
# 업데이트 후 인덱스 열 해제
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

중복된 책 이름의 행을 하나로 합치고 대출건수를 더했기 때문에 총 데이터는 줄어들고 대출건수 100회 이상인 책이 훨씬 늘었다.

```python
# 기존의 ns_book 데이터프레임의 열 이름을 그대로 전달
ns_book4 = ns_book4[ns_book.columns]
ns_book4.head()
```

# 일괄처리 함수 만들기 data_cleaning()

```python
def data_cleaning(filename):
    ns_df = pd.read_csv(filename, encoding='utf-8', low_memory=False)
    ns_book = ns_df.dropna(axis=1, how='all')
    count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
    loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
    dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN', '권'])
    unique_rows = ~dup_rows
    ns_book3 = ns_book[unique_rows].copy()
    ns_book3.set_index(['도서명','저자','ISBN', '권'], inplace = True)
    ns_book3.update(loan_count)
    ns_book4 = ns_book3.reset_index()
    ns_book4 = ns_book4[ns_book.columns]
    return ns_book4
```

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4)
```
