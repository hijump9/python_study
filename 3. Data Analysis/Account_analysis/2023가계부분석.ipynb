### 지출 내역 재분류
- 의료/건강 - 운동 을 문화/여가 - 스포츠 로 변경

```python
mask = (지출_df['대분류'] == '의료/건강') & (지출_df['소분류'] == '운동')
지출_df.loc[mask, '대분류'] = '문화/여가'
지출_df.loc[mask, '소분류'] = '스포츠'
```

### 결제수단
- 카드
    - KB국민 nori 체크카드(RF)          392   -> 노리 로 변경
    - KB국민 노리2 체크카드(KB Pay)_교통    363   -> 노리 2 로 변경
    - KB국민 그린체크카드(서울형)            160 -> 그린

- 통장
    - KB Star*t통장-저축예금            212 -> 국민스타트
    - 사이다입출금통장                     19 -> 사이다

- 간편결제
    - 카카오페이
        - 카카오페이 간편결제                   42 -> 카페 간편결제
        - 카카오페이 머니                     19 -> 카페 머니

    - 네이버페이
        - 네이버페이 간편결제(머니)               13 -> 네페 머니
        - 네이버페이 간편결제(포인트)              10 -> 네페 포인트
        - 네이버페이 간편결제                    2 -> 네페 간편결제

```python
# 새로운 열 '세부결제수단' 추가 및 값 설정
세부결제수단_매핑 = {
    'KB국민 nori 체크카드(RF)': '노리',
    'KB국민 노리2 체크카드(KB Pay)_교통': '노리 2',
    'KB국민 그린체크카드(서울형)': '그린',
    'KB Star*t통장-저축예금': '국민스타트',
    '사이다입출금통장': '사이다',
    'MY 입출금통장': '케이뱅크',
    '카카오페이 간편결제': '카페 간편결제',
    '카카오페이 머니': '카페 머니',
    '네이버페이 간편결제(머니)': '네페 머니',
    '네이버페이 간편결제(포인트)': '네페 포인트',
    '네이버페이 간편결제': '네페 간편결제'
}

# 세부결제수단 열 추가 및 매핑 적용
지출_df['세부결제수단'] = 지출_df['결제수단'].map(세부결제수단_매핑).fillna('기타')


# 결제수단 열을 '카드', '통장', '간편결제' 중 하나로 설정
지출_df['결제수단'] = 지출_df['결제수단'].map(lambda x: '카드' if '체크카드' in x else ('통장' if '통장' in x else '간편결제'))

# 열 순서 변경 - '메모' 열을 맨 뒤로 이동
열_순서 = ['날짜', '시간', '대분류', '소분류', '내용', '금액', '결제수단', '세부결제수단', '메모']
지출_df = 지출_df[열_순서]

지출_df
```
