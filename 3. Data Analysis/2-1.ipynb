## 파이썬에서 JSON 데이터 다루기

```python
d = {"name": "혼자 공부하는 데이터 분석"}

print(d['name'])
```

```python
# json.dumps() 함수는 파이썬 객체를 JSON 문자열로 변환한다.
import json
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
```

웹 기반 API는 전송하려는 파이썬 객체를 JSON 문자열로 변화해서 전송한다.
JSON 문자열을 파이썬 프로그램에 사용하려면 다시 파이썬 딕셔너리로 바꿔줘야 한다.

```python
# json.loads() 함수는 JSON 문자열을 파이썬 객체로 변환한다.
d2 = json.loads(d_str)

print(d2['name'])
```

### 직접 JSON 문자열을 json.loads()함수에 전달하기

```python
d3 = json.loads('{"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2022}')

print(d3['name'])
print(d3['author'])
print(d3['year'])
```

```python
d3 = json.loads('{"name": "혼자 공부하는 데이터 분석", "author": ["박해선","홍길동"], \
                  "year": 2022}')

print(d3['author'][1])
```

```python
d4_str = """
[
  {"name": "혼자 공부하는 데이터 분석",
  "author": "박해선",
  "year": 2022},
  {"name": "혼자 공부하는 머신러닝+딥러닝",
  "author": "박해선",
  "year": 2020}
]
"""
d4 = json.loads(d4_str)

print(d4[1]['name'])
```

```python
# JSON 문자열을 데이터프레임으로 변화하는 함수 : read_json() 함수
import pandas as pd
pd.read_json(d4_str)
```

## 파이썬에서 XML 데이터 다루기

```python
# XML 문자열을 파이썬 객체로 : fromstring() 함수
x_str = """
<book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et

book = et.fromstring(x_str)
```

```python
print(type(book)) # ElementTree 모듈 아래에 정의된 Element 클래스의 객체가 반환됨.
```

```python
book_childs = list(book) # 리스트로 변환

print(book_childs)
```

```python
# 이름, 저자, 연도를 변수에 할당 후 엘리먼트에 있는 텍스트를 출력
name, author, year = book_childs

print(name.text)
print(author.text)
print(year.text)
```

```python
# findtext() 메서드를 사용해 자식 엘리먼트를 탐색해서 자동으로 텍스트를 반환.
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
```

```python
x2_str = """
<books>
    <book>
        <name>혼자 공부하는 데이터 분석</name>
        <author>박해선</author>
        <year>2022</year>
    </book>
    <book>
        <name>혼자 공부하는 머신러닝+딥러닝</name>
        <author>박해선</author>
        <year>2020</year>
    </book>
</books>
"""
```

```python
# fromstring()함수를 사용해 파이썬 객체로 만들기
books = et.fromstring(x2_str)

print(books.tag)
```

```python
# 여러 개의 자식 엘리먼트를 확인하기 위해 findall() 메서드와 for문 사용
for book in books.findall('book'):
    name = book.findtext('name')
    author = book.findtext('author')
    year = book.findtext('year')

    print(name)
    print(author)
    print(year)
    print()
```
