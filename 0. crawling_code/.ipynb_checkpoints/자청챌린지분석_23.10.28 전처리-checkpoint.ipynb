```python
with open("./ìì²­ì±Œë¦°ì§€(9.29~10.6).txt", "r", encoding="utf-8") as file:
    chat_data = file.readlines()
    
chat_data[:10]
```

```python
# ë°ì´í„° ì €ì¥ì„ ìœ„í•œ ë¹ˆ ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
dates = []
days = []
names = []
times = []
urls = []

# ì„ì‹œ ë³€ìˆ˜ ì´ˆê¸°í™”
current_date = None
current_day = None
current_name = None
current_time = None

# ì •ê·œì‹ íŒ¨í„´ ì •ì˜
date_pattern = r'--------------- (\d{4}ë…„ \d{1,2}ì›” \d{1,2}ì¼) (.ìš”ì¼) ---------------'
name_pattern = r'\[(.*?)\]'
time_pattern = r'\[(ì˜¤ì „|ì˜¤í›„) (\d{1,2}:\d{2})\]'
url_pattern = r'http[s]?://\S+'

# ê° ì¤„ì„ ìˆœíšŒí•˜ë©° ë°ì´í„° ì¶”ì¶œ
for line in chat_data:
    date_match = re.search(date_pattern, line)
    name_match = re.search(name_pattern, line)
    time_match = re.search(time_pattern, line)
    url_match = re.findall(url_pattern, line)
    
    # ë‚ ì§œì™€ ìš”ì¼ ì •ë³´ ì—…ë°ì´íŠ¸
    if date_match:
        current_date = pd.to_datetime(date_match.group(1), format='%Yë…„ %mì›” %dì¼')
        current_day = date_match.group(2)
        continue
    
    # ì´ë¦„ ì •ë³´ ì—…ë°ì´íŠ¸
    if name_match:
        current_name = name_match.group(1)
    
    # ì‹œê°„ ì •ë³´ ì—…ë°ì´íŠ¸
    if time_match:
        am_pm, time = time_match.groups()
        hour, minute = map(int, time.split(':'))
        if am_pm == "ì˜¤í›„" and hour != 12:
            hour += 12
        current_time = f"{hour:02d}:{minute:02d}"
    
    # URL ì •ë³´ê°€ ìˆìœ¼ë©´ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
    if url_match:
        for url in url_match:
            dates.append(current_date)
            days.append(current_day)
            names.append(current_name)
            times.append(current_time)
            urls.append(url)

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame({
    'Date': dates,
    'Day': days,
    'Time': times,
    'Name': names,
    'URL': urls
})
df.head()
```

### [ID] ì—´ ì¶”ê°€

```python
def extract_id_from_url(row):
    url = row['URL']
    name = row['Name']
    
    # ë„¤ì´ë²„ ë¸”ë¡œê·¸ í˜•íƒœ (Postê°€ ë“¤ì–´ê°€ì§€ ì•ŠëŠ” ê²½ìš°)
    if "blog.naver.com" in url and "Post" not in url:
        return url.split('/')[3]
    # ë„¤ì´ë²„ ë¸”ë¡œê·¸ í˜•íƒœ (Postê°€ í¬í•¨ëœ ê²½ìš°)
    elif "blog.naver.com" in url or "m.blog.naver.com" in url:
        return re.search(r'blogId=([^&]*)', url).group(1)
    # í‹°ìŠ¤í† ë¦¬ í˜•íƒœ
    elif "tistory.com" in url:
        return urlparse(url).netloc.split('.')[0]
    # ê·¸ ì™¸ì˜ ê²½ìš°
    else:
        return name

df['ID'] = df.apply(extract_id_from_url, axis=1)
df.head()
```
