```python
with open("./자청챌린지(9.29~10.6).txt", "r", encoding="utf-8") as file:
    chat_data = file.readlines()
    
chat_data[:10]
```

```python
# 데이터 저장을 위한 빈 리스트 초기화
dates = []
days = []
names = []
times = []
urls = []

# 임시 변수 초기화
current_date = None
current_day = None
current_name = None
current_time = None

# 정규식 패턴 정의
date_pattern = r'--------------- (\d{4}년 \d{1,2}월 \d{1,2}일) (.요일) ---------------'
name_pattern = r'\[(.*?)\]'
time_pattern = r'\[(오전|오후) (\d{1,2}:\d{2})\]'
url_pattern = r'http[s]?://\S+'

# 각 줄을 순회하며 데이터 추출
for line in chat_data:
    date_match = re.search(date_pattern, line)
    name_match = re.search(name_pattern, line)
    time_match = re.search(time_pattern, line)
    url_match = re.findall(url_pattern, line)
    
    # 날짜와 요일 정보 업데이트
    if date_match:
        current_date = pd.to_datetime(date_match.group(1), format='%Y년 %m월 %d일')
        current_day = date_match.group(2)
        continue
    
    # 이름 정보 업데이트
    if name_match:
        current_name = name_match.group(1)
    
    # 시간 정보 업데이트
    if time_match:
        am_pm, time = time_match.groups()
        hour, minute = map(int, time.split(':'))
        if am_pm == "오후" and hour != 12:
            hour += 12
        current_time = f"{hour:02d}:{minute:02d}"
    
    # URL 정보가 있으면 데이터프레임에 추가
    if url_match:
        for url in url_match:
            dates.append(current_date)
            days.append(current_day)
            names.append(current_name)
            times.append(current_time)
            urls.append(url)

# 데이터프레임 생성
df = pd.DataFrame({
    'Date': dates,
    'Day': days,
    'Time': times,
    'Name': names,
    'URL': urls
})
df.head()
```

### [Format] 열추가

```python
def classify_format(url):
    if "blog.naver.com" in url or "m.blog.naver.co" in url:
        return "네이버"
    elif "www.instagram.com" in url:
        return "인스타"
    elif "tistory.com" in url:
        return "티스토리"
    elif "threads" in url:
        return "스레드"
    elif "youtu" in url:
        return "유튜브"
    elif "richmomjulie" in url:
        return "티스토리"
    else:
        return "그 외"
```

```python
df['Format'] = df['URL'].apply(classify_format)
df.head()
```

### [ID] 열 추가

```python
def extract_id_from_url(row):
    url = row['URL']
    name = row['Name']
    
    # 네이버 블로그 형태 (Post가 들어가지 않는 경우)
    if "blog.naver.com" in url and "Post" not in url:
        return url.split('/')[3]
    # 네이버 블로그 형태 (Post가 포함된 경우)
    elif "blog.naver.com" in url or "m.blog.naver.com" in url:
        return re.search(r'blogId=([^&]*)', url).group(1)
    # 티스토리 형태
    elif "tistory.com" in url:
        return urlparse(url).netloc.split('.')[0]
    # 그 외의 경우
    else:
        return name

df['ID'] = df.apply(extract_id_from_url, axis=1)
df.head()
```
